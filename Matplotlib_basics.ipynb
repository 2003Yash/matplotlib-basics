```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.random.random(50) *100 #first half gives an array of 50 random values between o and 1 and *100 is to bring the decimal nnumber int 2 steps left i.e 0.44 = 44
y_data = np.random.random(50) *100
```

**Scatter plots**

```python
plt.scatter(x_data,y_data)

#if graph is not showing then use plt.show()
```

```python
plt.scatter(x_data,y_data, c = "red") #to change color
```

```python
plt.scatter(x_data, y_data, c = "#0f0", marker = '*') #we can also specify coloor in hexdecimal and marker for each point
```

```python
a_data = np.random.random(1000) *100
b_data = np.random.random(1000) *100

plt.scatter(a_data, b_data, c = "#00f", alpha = 0.3)
#aplha is just opacity and if points overlap we can see darker areas if alpha = 0.9 all are darker areas since every point has full opacity and vice versa
```

Line plots

```python
years = [2006 + x for x in range(16)]
weights = [88,84,86,86, 90, 92, 82, 93, 83, 94, 85, 83, 84, 92, 85, 92]
years
```

```python
plt.plot(years, weights) #by default it uses line plot
```

```python
plt.plot(years, weights, c = 'r', lw = 3) #r describes red and lw means line weight to get thicker line
```

```python
plt.plot(years, weights, c = 'r', linestyle = "--")
# or plt.plot(years, weights, c = 'r--')
```

Bar plots

```python
x = ["c++", "c#", "python", "java"]
y = [20, 25,67,23]

plt.bar(x,y)
```

```python
plt.bar(x,y, color = 'r', align = "edge")
#we have to use color not c
#align will start the graph from point of it's catergory check diff wrt to x axis
```

```python
plt.bar(x,y, color = 'r', align = "edge", width = 0.3, edgecolor = 'g') #width describes the bars width edgecolour describes edge color of each bar
```

Histograms

```python
ages = np.random.normal(20,1.5,1000) # creates normal distribution with 20 mean and sd of 1.5 and 100 different random vales

plt.hist(ages)
```

```python
plt.hist(ages, bins=2) #bins is no.of bars to describe the historgrams
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])  #here we are creating a seperate biuns starting to 18 age and a18 to 21 and from 21 till end
```

```python
plt.hist(ages, bins=20, cumulative = True) #to plot cumulative dist function
```

Pie-Charts

```python
x = ["c++", "c#", "python", "java"]
y = [20, 25,67,23]

plt.pie(y,labels = x)
```

```python
explodes = [0,0,0.2,0]

plt.pie(y, labels = x, explode = explodes)
#this will highlight a seperate area in pir=e chart
```

```python
plt.pie(y, labels = x, explode = explodes, autopct = "%.2f%%") # to show percentages upto 2 decimal points AUTOPCT means auto percentage
```

```python
plt.pie(y, labels = x, explode = explodes, autopct = "%.2f%%", pctdistance = 0.9)  #to move percentage away from axis
```

Box-plots

```python
height = np.random.normal(172,8,300)
plt.boxplot(height)
```

To understand Box plot:
1- yellow line describes median
2- box above edge is 3rd quartile and box below edge is 1st quartile
3- overall edges of graph shows min and max values in dist
4- no. of values in ablove region of box is euqal to first half  and second half of bax and below region of box
5- outliers are points that lie outside the range of box plot
